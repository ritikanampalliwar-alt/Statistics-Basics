# Statistics Basics

```python
#1. What is the difference between descriptive statistics and inferential
# statistics? Explain with examples.
# Ans:- 1. Descriptive Statistics:

# Definition:
# Descriptive statistics summarize and describe the features of a dataset.

# Purpose:
# To organize, simplify, and present data in a meaningful way.

# Common Methods:
#  Mean, Median, Mode
#  Range, Variance, Standard Deviation
#  Tables, Graphs (bar charts, histograms)

# Example:

# You surveyed 100 students and found that the average (mean) height is 160 cm.
# → This is just describing your collected data.

# 2. Inferential Statistics:

# Definition:
# Inferential statistics use a sample of data to make predictions, decisions, or conclusions about a larger population.

# Purpose:
# To infer something beyond the observed data.

# Common Methods:
# Hypothesis Testing (t-test, z-test, chi-square test)
# Confidence Intervals

# Example:
# You collected height data from 100 students, and then you predict or conclude the average height of all students in the college.
# → You are not measuring everyone—you are inferring about the population.



# 2.What is sampling in statistics? Explain the differences between random
# and stratified sampling.
# Ans:- Sampling is the process of selecting a small group (sample) from a large group (population) so that we can study the small group and make conclusions about the whole population.
# Example:
# A college has 5,000 students.
# You want to know the average height.
# Measuring all 5,000 students is difficult.
# So you measure only 200 students → This 200 is your sample.

# 1. Random Sampling (Simple Random Sampling):Every individual in the population has an equal chance of being selected.
# Example:
# A college has 5,000 students.
# You randomly pick 200 roll numbers using a computer.Everyone had an equal chance to be selected.

# 2.Stratified Sampling:Population is divided into subgroups (strata) based on a characteristic like:
# Gender
# Age
# Income
# Then we take samples from each subgroup to ensure representation.

# Example:College has:
# 3,000 boys
# 2,000 girls
# You want 200 students total.
# So take samples in proportion:
# From boys: (3000/5000)*200 = 120
# From girls: (2000/5000)*200 = 80
# Both gender groups are represented fairly.



# 3.Define mean, median, and mode. Explain why these measures of central tendency are important.
# 1. Mean (Average)
# Definition:Mean is the sum of all values divided by the total number of values.

# Example:
# Numbers: 10, 20, 30
# Mean = (10 + 20 + 30) ÷ 3 = 20

# 2. Median
# Definition:Median is the middle value when numbers are arranged in ascending order.

# Example:Numbers: 5, 8, 12 → Median = 8
# If even count: Median = average of two middle values.

# 3. Mode
# Definition:Mode is the value that appears most frequently in the data.
# Example:Numbers: 2, 3, 3, 5 → Mode = 3

# Why These Measures of Central Tendency Are Important?

# These measures help us understand the center, typical value, or common pattern of a dataset.
# 1. They summarize data in one number:Instead of looking at 1,000 values, mean/median/mode give a quick idea about the data.
# 2. They help in comparing datasets
# For example, comparing average salaries of two companies.
# 3. They describe the central behavior of data:Mean shows overall level, median shows middle point, and mode shows most common value.
# 4. Useful in decision making:
# Mean is used for average marks, income, temperature.
# Median is useful when data has outliers (income distribution).
# Mode is used in business, marketing, and when most frequent value matters.



# 4.Explain skewness and kurtosis. What does a positive skew imply about the
# data?
# Answer:1. Skewness
# Definition:Skewness measures the asymmetry (tilt) of a distribution.
# If data is perfectly symmetric, skewness = 0
# If data is not symmetric, it is skewed.

# Types of Skewness:
# Positive Skew (Right Skew)
# Negative Skew (Left Skew)

# 2. Kurtosis
# Definition:Kurtosis measures the tailedness of the distribution —
# how heavy or light the tails are compared to a normal distribution.

# Types of Kurtosis:
# Leptokurtic: High kurtosis, heavy tails (more extreme values)
# Mesokurtic: Normal kurtosis
# Platykurtic: Low kurtosis, light tails (fewer extreme values)

# What Does a Positive Skew imply about the data:Positive skew means the tail of the distribution extends to the right side.

# Interpretation:
# Most values are towards the left (lower values)
# Few very large values pull the tail towards the right
# Mean > Median > Mode

# Real-life examples of positive skew:
# Income distribution
# House prices
```

```python
# 5. Implement a Python program to compute the mean, median, and mode of
# a given list of numbers.
# numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
# Answer:
from statistics import mean, median, mode

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate mean, median, mode
mean_value = mean(numbers)
median_value = median(numbers)
mode_value = mode(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

```python
# 6.Question 6: Compute the covariance and correlation coefficient between the following
# two datasets provided as lists in Python:
# list_x = [10, 20, 30, 40, 50]
# list_y = [15, 25, 35, 45, 60]
# Answer:
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

x = np.array(list_x)
y = np.array(list_y)

cov_matrix = np.cov(x, y)     # sample covariance
cov_xy = cov_matrix[0, 1]
corr_xy = np.corrcoef(x, y)[0, 1]

print("Covariance:", cov_xy)
print("Correlation Coefficient:", corr_xy)

```

```python
# 7.Question 7: Write a Python script to draw a boxplot for the following numeric list and
# identify its outliers. Explain the result:
# data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
# Answer:
import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Plot boxplot
plt.boxplot(data)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

outliers

# Explanation of Result:
# 1. IQR (Interquartile Range) Method:
# Q1 (25th percentile) = 18
# Q3 (75th percentile) = 24
# IQR = Q3 – Q1 = 24 – 18 = 6

# 2. Outlier Boundaries:
# Lower bound = Q1 – 1.5×IQR = 18 – 9 = 9
# Upper bound = Q3 + 1.5×IQR = 24 + 9 = 33

# 3. Identifying Outliers:
# Any number < 9 or > 33 is an outlier.
```

```python
# 8. You are working as a data analyst in an e-commerce company. The
# marketing team wants to know if there is a relationship between advertising spend and
# daily sales.
# ● Explain how you would use covariance and correlation to explore this
# relationship.
# ● Write Python code to compute the correlation between the two lists:
# advertising_spend = [200, 250, 300, 400, 500]
# daily_sales = [2200, 2450, 2750, 3200, 4000]
# Answer:1. Covariance:Covariance tells you the direction of the relationship.

# Positive covariance → When ad spend increases, sales also increase
# Negative covariance → When ad spend increases, sales decrease
# Covariance close to 0 → No relationship
# However, covariance does not tell the strength of the relationship (only direction).

# ✔ 2. Correlation:Correlation tells you:

# Direction (positive/negative)
# Strength of the relationship
# Value is always between –1 and +1

# Interpretation:
# +1 → Perfect positive relationship
# 0 → No relationship
# –1 → Perfect negative relationship
# Correlation is more useful for decision-making because it is standardized and easy to interpret.
    
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

x = np.array(advertising_spend)
y = np.array(daily_sales)

correlation = np.corrcoef(x, y)[0, 1]

print("Correlation between advertising spend and daily sales:", correlation)
```

```python
# 9.Your team has collected customer satisfaction survey data on a scale of
# 1-10 and wants to understand its distribution before launching a new product.
# ● Explain which summary statistics and visualizations (e.g. mean, standard
# deviation, histogram) you’d use.
# ● Write Python code to create a histogram using Matplotlib for the survey data:
# survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
# Answer:To understand the distribution of customer satisfaction scores (1–10 scale), I would use the following:
# 1. Summary Statistics
# Mean – tells the average satisfaction level.
# Median – helps understand the central tendency and reduces the effect of outliers.
# Mode – shows the most common rating.
# Standard Deviation – indicates how spread out the scores are.
# Minimum & Maximum – show the range of responses.
# Skewness – shows if responses lean more toward low or high scores.
# Kurtosis – tells whether the distribution is peaked or flat.

# 2. Visualizations
# Histogram – best for checking the overall distribution.
# Boxplot – helpful for identifying outliers.
# Bar Chart – useful since the data is discrete (1–10 scores).
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.hist(survey_scores, bins=6, edgecolor='black')
plt.xlabel('Survey Score')
plt.ylabel('Frequency')
plt.title('Customer Satisfaction Score Distribution')
plt.show()

```
